```python
import pandas as pd

# Set up a dictionary to store unique values and their counts
unique_values = {}

# Define a function to process a chunk of data
def process_chunk(chunk):
    # Get the third column, you can also use chunk.iloc[:, 2] if you prefer
    col = chunk.iloc[:, 2]

    # Count unique values and their occurrences in the chunk
    chunk_counts = col.value_counts()

    # Update the global unique_values dictionary with the chunk_counts
    for value, count in chunk_counts.items():
        if value in unique_values:
            unique_values[value] += count
        else:
            unique_values[value] = count

# Read and process the CSV file in chunks
chunksize = 10**6  # You can adjust this value depending on your memory constraints
file_path = r'C:\Users\nuke2\Desktop\NW Work\Spring Work\MSiA-SQ\Data\MSiA 431\02_hw\music_sample.csv'
for chunk in pd.read_csv(file_path, chunksize=chunksize, low_memory=False):
    process_chunk(chunk)

# Print the unique values and their counts
print(unique_values)
print("Number of unique values in the third column: ", len(unique_values))
```
